# 2 read .csv from multiple URLs with pandas
https://www.football-data.co.uk/mmz4281/2425/E0.csv
# Link:root+season+leauge

```python
# link structure
"https://www.football-data.co.uk/mmz4281/"+"2425"+"/"+"E0"+".csv"
```

```python
# create a root variable
root="https://www.football-data.co.uk/mmz4281/"
```

```python
# creating list of leauges
leauges=['E0','E1','E2','E3','EC']
frames=[]
#  looping through leauges, read multiple csv and append it into a list
for leauge in leauges:
    df=pd.read_csv(root+'2425'+'/'+leauge+'.csv')
    frames.append(df)
```

```python
# length of frames
len(frames)
```

```python
# show 1st 2nd 3rd element
frames[4]
```

# 2.2 multiple seasons

```python
for season in range(15,25):
    print(str(season)+str(season+1))
```

```python
# loop through range of seasons and insert new coloumn
leauges=['E0','E2','E3']
frames=[]
#  looping through leauges, read multiple csv and append it into a list
for leauge in leauges:
    for season in range(15,25):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+leauge+'.csv')
        df.insert(1,'season',season)
        frames.append(df)
```

```python
# total frames(10 seasons x 3 leagues)
len(frames)
```

```python
# show 1st and last element (new season coloumn was created)
frames[29]
```

# organize all the data in a dictonary

```python
# create a dictonary with origional name of leauges as key
dict_countries={
               'Spainish La Liga':'SP1','Spainish Segunda Division':'SP2',
               'German Bundesliga':'D1',
               'English Premier Leauge':'E0','English Leauge 1':'E2',
               'English Leauge 2':'E3'
}
```

```python
# get a dictionary element
dict_countries['Spainish La Liga']
```

```python
for leauge in dict_countries:
    print(dict_countries[leauge])
```

```python
dict_historical_data={}
#leauges=['E0','E2','E3']

#  looping through key elements
for leauge in dict_countries:
    frames=[]
    for season in range(15,25):
        df=pd.read_csv(root+str(season)+str(season+1)+'/'+dict_countries[leauge]+'.csv')
        df.insert(1,'season',season)
        frames.append(df)
    df_concat=pd.concat(frames)
    dict_historical_data[leauge]=df_concat
```
